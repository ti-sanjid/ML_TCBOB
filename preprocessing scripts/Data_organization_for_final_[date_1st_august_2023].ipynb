# Cyclone Ellimination Function


```python

def filter_cyclone_data(cyclone_data,min_lifetime):
    # Convert 'ISO_TIME' column to datetime
    cyclone_data['ISO_TIME'] = pd.to_datetime(cyclone_data['ISO_TIME'])

    # Group the DataFrame by 'NUMBER'
    grouped_data = cyclone_data.groupby('NUMBER')

    # Calculate the total hours passed for each group
    hours_passed = grouped_data['ISO_TIME'].apply(lambda x: (x.max() - x.min()).total_seconds() / 3600)

    # Filter the DataFrame based on the condition
    filtered_data = cyclone_data[cyclone_data['NUMBER'].isin(hours_passed[hours_passed >= min_lifetime].index)]

    return filtered_data

# give df and minimum life span of a cyclone in function
df=filter_cyclone_data(df,48)
```

# Lat +6

```python
def lat_plus_n(predict_hour,data):
  steps=(predict_hour/3)
  # Create a new column 'Lat +n' as an exact copy of 'LAT'
  col_name='Lat +'+ str(predict_hour)
  data[col_name] = data['lat']
  # Group the DataFrame by 'NUMBER'
  grouped_data = data.groupby('NUMBER')
  # Shift the values within each group by the value steps
  data[col_name] = grouped_data[col_name].shift(-steps)

lat_plus_n(6,df)
```

# Lon +6

```python
def lon_plus_n(predict_hour,data):
  steps=(predict_hour/3)
  # Create a new column 'Lon +n' as an exact copy of 'LON'
  col_name='Lon +'+ str(predict_hour)
  data[col_name] = data['lon']
  # Group the DataFrame by 'NUMBER'
  grouped_data = data.groupby('NUMBER')
  # Shift the values within each group by the value steps
  data[col_name] = grouped_data[col_name].shift(-steps)

lon_plus_n(6,df)
```
