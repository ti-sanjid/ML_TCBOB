## Cyclone Ellimination Function(48hrs)

```python
def filter_cyclone_data(cyclone_data,min_lifetime):
    # Convert 'ISO_TIME' column to datetime
    cyclone_data['Date'] = pd.to_datetime(cyclone_data['Date'])

    # Group the DataFrame by 'NUMBER'
    grouped_data = cyclone_data.groupby('cyclone_no')

    # Calculate the total hours passed for each group
    hours_passed = grouped_data['Date'].apply(lambda x: (x.max() - x.min()).total_seconds() / 3600)

    # Filter the DataFrame based on the condition
    filtered_data = cyclone_data[cyclone_data['cyclone_no'].isin(hours_passed[hours_passed >= min_lifetime].index)]

    return filtered_data

# give df and minimum life span of a cyclone in function
data=filter_cyclone_data(data,48)
```

# 3.5 degree rule

```python
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Define a function to calculate Haversine distance (same as before)
def haversine_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine formula (same as before)
    d_lat = lat2 - lat1
    d_lon = lon2 - lon1
    a = sin(d_lat/2)**2 + cos(lat1) * cos(lat2) * sin(d_lon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    earth_radius_km = 6371.0  # Earth's radius in kilometers
    distance_km = earth_radius_km * c

    return distance_km



# New_df = pd.DataFrame(columns=df.columns)



def filtered_cyclone(df,cyclone_number):
  grouped_data = df.groupby('cyclone_no')
  specific_cyclone_data = grouped_data.get_group(cyclone_number)

  for i in range(len(specific_cyclone_data) - 1):
    lat1, lon1 = specific_cyclone_data.at[specific_cyclone_data.index[i],'lat'], specific_cyclone_data.at[specific_cyclone_data.index[i],'lon']
    lat2, lon2 = specific_cyclone_data.at[specific_cyclone_data.index[i+1], 'lat'], specific_cyclone_data.at[specific_cyclone_data.index[i+1], 'lon']
    # print(lat1,lon1,lat2,lon2)
    row_index_exceeds_350km=None

    distance = haversine_distance(lat1, lon1, lat2, lon2)
    if distance >= 300:
      row_index_exceeds_350km=i+1
      break
  specific_cyclone_data_filtered=specific_cyclone_data.iloc[:row_index_exceeds_350km]
  return specific_cyclone_data_filtered




# Get the unique cyclone numbers in the DataFrame
unique_cyclone_numbers = data['cyclone_no'].unique()
New_df = pd.DataFrame(columns=data.columns)
# Iterate through each unique cyclone number and get its data
for cyclone_number in unique_cyclone_numbers:
    specific_cyclone_data = filtered_cyclone(data, cyclone_number)
    New_df = pd.concat([New_df, specific_cyclone_data], ignore_index=True)
```

```python
import pandas as pd

def calculate_monthly_counts(dataframe, group_column, datetime_column):
    # Group the DataFrame by the specified column
    grouped = dataframe.groupby(group_column)

    # Initialize a dictionary to store monthly counts
    monthly_counts = {}

    # Iterate through each group
    for _, group_data in grouped:
        # Get the unique months in the current group
        group_unique_months = group_data[datetime_column].str[5:7].unique()

        # Update the overall monthly counts dictionary
        for month in group_unique_months:
            if month in monthly_counts:
                monthly_counts[month] += 1
            else:
                monthly_counts[month] = 1

    return monthly_counts

# Assuming your DataFrame is named 'df2'
group_column = 'cyclone_no'
datetime_column = 'Date'
monthly_counts = calculate_monthly_counts(data, group_column, datetime_column)


monthly_counts
```

```python
month_name = {
    '01': 'January',
    '02': 'February',
    '03': 'March',
    '04': 'April',
    '05': 'May',
    '06': 'June',
    '07': 'July',
    '08': 'August',
    '09': 'September',
    '10': 'October',
    '11': 'November',
    '12': 'December'
}
month_name = {value:key for key,value in month_name.items()}
month_name
```

```python
monthly_cyclone_count={}
for month_key, month_value in month_name.items():
    if month_value in monthly_counts:
        monthly_cyclone_count[month_key] = monthly_counts[month_value]
    else:
        monthly_cyclone_count[month_key] = 0
```

```python
# Convert the result dictionary to a DataFrame
monthly_cyclone_count_df = pd.DataFrame(monthly_cyclone_count.items(), columns=['Month', 'Occurrences'])
monthly_cyclone_count_df
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))

colors = ['magenta', 'magenta', 'orange', 'orange', 'orange', 'purple', 'purple', 'purple', 'purple', 'yellow', 'yellow', 'magenta']
plt.bar(monthly_cyclone_count_df['Month'], monthly_cyclone_count_df['Occurrences'],color=colors)
plt.xlabel('Month')
plt.ylabel('Occurrences')
plt.title('Monthly occurrences of cyclones')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
print('-----------------------------------------')
print('Premoonson---orange,Moonsoon-----purple,Postmoonsoon-yellow')
```

## Yearly cyclone count

```python
import pandas as pd

def calculate_yearly_counts(dataframe, group_column, datetime_column):
    # Group the DataFrame by the specified column
    grouped = dataframe.groupby(group_column)

    # Initialize a dictionary to store monthly counts
    yearly_counts = {}

    # Iterate through each group
    for _, group_data in grouped:
        # Get the unique months in the current group
        group_unique_year = group_data[datetime_column].str[:4].unique()

        # Update the overall monthly counts dictionary
        for year in group_unique_year:
            if year in yearly_counts:
                yearly_counts[year] += 1
            else:
                yearly_counts[year] = 1

    return yearly_counts

# Assuming your DataFrame is named 'df2'
group_column = 'cyclone_no'
datetime_column = 'Date'
yearly_counts = calculate_yearly_counts(data, group_column, datetime_column)


yearly_counts
```

```python
yearly_cyclone_count_df = pd.DataFrame(yearly_counts.items(), columns=['Year', 'Occurrences'])
yearly_cyclone_count_df
```
